```python
from skimage import io
import matplotlib.pyplot as plt

imgColor = io.imread("hoja.jpg")
imgColor.shape
```

```python
plt.figure() 
plt.imshow (imgColor) 
plt.show()
```

```python
C = ['Reds', 'Greens', 'Blues']
for p in [0,1,2]:
    plt.figure() 
    plt.imshow (imgColor[:,:,p],cmap=C[p]) 
    plt.show()
```
