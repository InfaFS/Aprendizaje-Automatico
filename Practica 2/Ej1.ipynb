```python
import pandas as pd
import numpy as np
from matplotlib import pylab as plt

from sklearn import model_selection,preprocessing

from ClassPerceptron import Perceptron
```

```python
data = {
    'TEMPERATURA' : [-12,20,0,7,11,-30,63,94,45],
    'DUREZA': [60,30,60,30,0,60,0,0,30],
    'RESISTENCIA' : [1,1,1,1,0,1,0,0,0],
}

df = pd.DataFrame(data)
print(df)

#Datos de entrenamiento
X = np.array(df.iloc[:,0:2])
print(X)
#Salida
T = df['RESISTENCIA']
T = np.array(T)
print(T)
```

```python
ppn = Perceptron(alpha=0.001, n_iter=350, draw=1, title='Materiales', random_state=None)
ppn.fit(X, T)
```

```python
Y = ppn.predict(X)
print("Y = ",Y)
print ("T = ", T)
aciertos = sum(Y == T)
print("Aciertos = ",aciertos)
print("%% de aciertos = %.2f %%" % (100*aciertos/X.shape[0]))
```

```python
X_test = np.array([[9,0]])
print(X_test)
Y = ppn.predict(X_test)
print(Y)
```
