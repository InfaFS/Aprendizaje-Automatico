```python
import pandas as pd
import numpy as np
from matplotlib import pylab as plt

from sklearn import model_selection,preprocessing

from ClassPerceptron import Perceptron
```

```python
data = pd.read_csv("zoo.csv", encoding="latin1") 
# print(data)

#Vamos a modificar la data para que solo tenga 1 si es mamifero o 0 caso contrario
#Cambiar para las aves y para los reptiles
data['Clase'] = data['Clase'].replace({
    'Mamifero': 0,
    'Ave' : 0,
    'Pez' : 0,
    'Invertebrado' : 0,
    'Insecto': 0,
    'Reptil': 1,
    'Anfibio':0,
})
# print(data)
nColum = list(data.columns.values)
T_Train = data['Clase'].values
X_Train = data.iloc[:,1:18].values

T_Test = data['Clase'].values
X_Test = data.iloc[:,1:18].values

normalizador= preprocessing.StandardScaler()
    
X_Train = normalizador.fit_transform(X_Train)
X_Test  = normalizador.fit_transform(X_Test)

```

```python
ppn = Perceptron(alpha=0.01, n_iter=200,draw=1 ,random_state=None)
ppn.fit(X_Train, T_Train)

plt.plot(range(1, len(ppn.errors_) + 1), ppn.errors_, marker='o')
plt.xlabel('Iteraciones')
plt.ylabel('Cantidad de actualizaciones')
plt.show()
```

```python
Y_test = ppn.predict(X_Test)

aciertos = sum(Y_test == T_Test)
print("aciertos = ", aciertos)

nAciertos = sum(Y_test == T_Test)
print("%% de aciertos = %.2f %%" % (100*nAciertos/X_Test.shape[0]))

print(Y_test)
print(T_Test)

```
