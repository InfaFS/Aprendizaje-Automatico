```python
import numpy as np
import math
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import model_selection,preprocessing,metrics
import sys
sys.path.append("../Recursos")
from ClassNeuronaLineal import NeuronaLineal
```

```python
data = pd.read_csv("../Datos/automobile-simple.csv")
price_mean = data["price"].mean()

# Reemplazar los valores vacíos en la columna 'price' con la media
data['price'].fillna(price_mean, inplace=True)

# Verificar que los valores vacíos han sido reemplazados
# print(data['price'].isnull().sum())  # Debería mostrar 0

X = data['engine-size'].values.reshape(-1,1)
Y = data['price'].values

#sin normalizar
# normalizador = preprocessing.MinMaxScaler()
normalizador = preprocessing.StandardScaler()
X = normalizador.fit_transform(X)
Y = normalizador.fit_transform(Y.reshape(-1,1)) #Normalizamos porque no es un valor 0 o 1, luego para predecir debemos invertir la normalizacion si queremos
#Por ejemplo si tenemos una entrada como antes, la normalizamos, predecimos y luego invertimos la normalizacion para que muestre un valor acorde al esperado
print(X)
```

```python
combinador = NeuronaLineal(alpha=0.001, n_iter=100, cotaE=10e-06, draw=1, title=['X', 'Y'])
combinador.fit(X,Y)
```
