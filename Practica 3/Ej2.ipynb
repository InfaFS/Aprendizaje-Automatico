```python
import numpy as np
import math
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
#Ejercicio a)
#Vamos a simular la convergencia 

#Funcion principal
r = 1
h = 1
def f(r,h):
    return (r**2*h*math.pi/3) 

#Derivadas
df_r = (2*r*h*math.pi/3)
df_h = (r**2*math.pi/3)

MAX_ITE = 1000
learning_rate = 0.01

for i in range (MAX_ITE):
    df_r_temp = (2*r*h*math.pi/3)
    df_h_temp = (r**2*math.pi/3)
    r = r - (learning_rate * df_r_temp)
    h = h - (learning_rate * df_h_temp)

print("r = " ,r)
print("h = ",h)
print("valor funcion con los 2 valores = ",f(r,h))
```

```python

# Definir la funcin
def f(r, h):
    return (r**2 * h * np.pi) / 3

# Crear una malla de valores de r y h
r = np.linspace(-5, 5, 100)
h = np.linspace(-5, 5, 100)
R, H = np.meshgrid(r, h)
Z = f(R, H)

# Crear el grfico 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Dibujar la superficie
ax.plot_surface(R, H, Z, cmap='viridis')

# Etiquetas y ttulo
ax.set_xlabel('r')
ax.set_ylabel('h')
ax.set_zlabel('f(r, h)')
ax.set_title('Grfico 3D de la funcin f(r, h)')

# Mostrar el grfico
plt.show()
```

```python
#b)

#Funcion principal
x = 1
y = 1
def f(x,y):
    return (-3/(x**2+y**2+1)) 
#En este caso apuntamos al -3 (ya que si z es = -3 obtuvimos el minimo)


MAX_ITE = 1000
learning_rate = 0.01

for i in range (MAX_ITE):
    df_x_temp = (6*x/(x**2+y**2+1)**2)
    df_y_temp = (6*y/(x**2+y**2+1)**2)
    x = x - (learning_rate * df_x_temp)
    y = y - (learning_rate * df_y_temp)

print("x = " ,x)
print("y = ",y)
print("valor funcion con los 2 valores = ",f(x,y))
```
